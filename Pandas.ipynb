Heres a **clear, structured feature description** for each of those Titanic dataset columns:

---

### **1. survived**

* **Type:** Integer (0 or 1)
* **Meaning:** Survival status of the passenger.

  * `0` = Did not survive
  * `1` = Survived
* **Role:** **Target variable** for prediction tasks.

---

### **2. pclass**

* **Type:** Integer (1, 2, or 3)
* **Meaning:** Passenger class (a proxy for socio-economic status).

  * `1` = First class (highest)
  * `2` = Second class
  * `3` = Third class (lowest)
* **Notes:** Lower class often had lower survival rates.

---

### **3. sex**

* **Type:** Categorical (`male` / `female`)
* **Meaning:** Gender of the passenger.

---

### **4. age**

* **Type:** Float (in years)
* **Meaning:** Age of the passenger.
* **Special Cases:** May contain missing values.

---

### **5. sibsp**

* **Type:** Integer
* **Meaning:** Number of siblings or spouses aboard the Titanic with the passenger.
* **Notes:** Helps identify family groups.

---

### **6. parch**

* **Type:** Integer
* **Meaning:** Number of parents or children aboard the Titanic with the passenger.

---

### **7. fare**

* **Type:** Float
* **Meaning:** Ticket fare paid by the passenger (in British pounds).
* **Notes:** Varies by `pclass` and cabin type.

---

### **8. embarked**

* **Type:** Categorical (`C`, `Q`, `S`)
* **Meaning:** Port of embarkation.

  * `C` = Cherbourg, France
  * `Q` = Queenstown, Ireland
  * `S` = Southampton, England
* **Special Cases:** May contain missing values.

---

### **9. class**

* **Type:** Categorical (`First`, `Second`, `Third`)
* **Meaning:** Human-readable version of `pclass`.

---

### **10. who**

* **Type:** Categorical (`man`, `woman`, `child`)
* **Meaning:** Grouped category based on `age` and `sex`.

---

### **11. adult\_male**

* **Type:** Boolean (`True` / `False`)
* **Meaning:** Whether the passenger is an adult male.
* **Rule:** `True` if `sex` = male and `age` >= \~16 years.

---

### **12. deck**

* **Type:** Categorical (AG, NaN)
* **Meaning:** Passenger's cabin deck letter.
* **Special Cases:** Many missing values (unknown cabins).

---

### **13. embark\_town**

* **Type:** Categorical
* **Meaning:** Name of embarkation town.

  * Cherbourg, Queenstown, Southampton.

---

### **14. alive**

* **Type:** Categorical (`yes` / `no`)
* **Meaning:** Text version of survival status.

---

### **15. alone**

* **Type:** Boolean (`True` / `False`)
* **Meaning:** Whether the passenger was traveling alone.
* **Rule:** `True` if `sibsp` + `parch` = 0.

---
